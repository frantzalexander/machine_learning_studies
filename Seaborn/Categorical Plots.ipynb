```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/tips.csv")
```

```python
df.info()
```

```python
df["sex"].value_counts()
```

```python
df["size"].value_counts()
```

```python
plt.figure(figsize = (4,4), dpi = 150)
sns.countplot(
    data = df,
    x = "size"
);
```

```python
sns.countplot(
    data = df,
    x = "day"
);
```

```python
sns.countplot(data = df, x = "time");
```

```python
sns.barplot(
    data = df, 
    x = "time", 
    y = "total_bill", 
    estimator= np.mean, 
    errorbar = "sd",
    palette= "");
```

```python
sns.barplot(
    data = df,
    x = "size",
    y = "total_bill",
    hue = "sex",
    estimator = np.mean,
    errorbar = "sd",
)
plt.legend(bbox_to_anchor = (1.05, 1));
```
