```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
tips = pd.read_csv("../DATA/tips.csv")
tips.head()
```

```python
tips_corr = tips.drop("CC Number", axis = 1).corr(numeric_only= True)
tips_corr
```

```python
sns.heatmap(
    tips_corr, 
    annot= True, 
    linewidths= 0.7, 
    cmap = "flare",
    linecolor= "black",
    center = 0.8
    );
```

```python
flights.pivot_table(
    values = "passengers",
    index = "month",
    columns = "year",
    observed = True
)
```

```python
plt.figure(figsize= (15, 8))
flights_pivot = flights.pivot_table(
    values = "passengers",
    index = "month",
    columns = "year",
    observed = True
)

sns.heatmap(flights_pivot, cmap = "magma");
```

```python
sns.clustermap(
    tips_corr, 
    annot= True, 
    linewidths= 0.7, 
    cmap = "flare",
    linecolor= "black",
    col_cluster = False
    );
```
