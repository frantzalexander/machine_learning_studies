```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/tips.csv")
```

```python
df.head()
```

## Catplot

```python
sns.catplot(
    data = df,
    x = "sex",
    y = "total_bill",
    kind = "box",
);
```

```python
sns.catplot(
    data = df,
    x = "day",
    y = "total_bill",
    kind = "box",
    row = "smoker",
    col = "sex"
);
```

---

## Pairgrid

```python
sns.pairplot(
    data = df
);
```

```python
g = sns.PairGrid(
    data = df,
    hue = "sex"
    )
g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend();
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=df)
g = g.plot(sns.regplot, sns.histplot)
```
