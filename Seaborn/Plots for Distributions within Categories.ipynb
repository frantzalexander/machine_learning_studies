```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/iris.csv")
```

```python
df.head()
```

```python
sns.boxplot(
    data = df, 
    y = "sepal_length", 
    x = "species", 
    hue = "species", 
    palette= "Set2"
    );
```

```python
sns.violinplot(
    data = df, 
    y = "species", 
    x = "sepal_length", 
    hue = "species", 
    palette= "Set2",
    orient= "h"
    );
```

```python
df_tips = pd.read_csv("../DATA/tips.csv")
```

```python
df_tips.info()
```

```python
sns.boxplot(
    data = df_tips,
    x = "day",
    y = "total_bill",
    hue = "smoker",
    palette= "Set1"
    );
```

```python
plt.figure(figsize= (8, 6), dpi = 200)
sns.violinplot(
    data = df_tips,
    x = "total_bill",
    y = "day",
    orient= "h",
    hue = "sex",
    inner = None,
    split = True
);
```

```python
plt.figure(figsize= (8, 6), dpi = 200)
sns.violinplot(
    data = df_tips,
    x = "total_bill",
    y = "day",
    orient= "h",
    hue = "sex",
    inner = "stick",
    split = True
)
plt.legend(bbox_to_anchor = (1.2, 1));
```

```python
plt.figure(figsize= (8, 4), dpi = 200)
sns.swarmplot(
    data = df_tips, 
    x = "total_bill",
    y = "sex", 
    hue = "smoker",
    dodge = True, 
    palette= "Set1"
    );
```

```python
sns.boxenplot(
    data = df_tips,
    x = "total_bill",
    y = "sex",
    hue = "smoker",
    palette= "mako"
);
```
