```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/Telco-Customer-Churn.csv")
```

```python
df.info()
```

```python
df["InternetService"].unique()
```

```python
plt.figure(figsize = (6,6))
sns.scatterplot(
    data = df.sample(frac = 0.1), 
    x = "MonthlyCharges",
    y = "TotalCharges",
    hue = "InternetService",
    s = 100,
    alpha = 0.5,
    palette = "flare",
    style = "InternetService")

plt.savefig("Scatterplot of Customer charges by Gender");
```

# Distribution Plots

```python
data = pd.read_csv("../../metropt+3+dataset/MetroPT3(AirCompressor).csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data["Oil_temperature"]
```

```python
plt.figure(figsize = (5,5), dpi = 150)

sns.rugplot(
    x = "Oil_temperature", 
    data = data.sample(frac = 0.1), 
    height = 0.02
    );
```

```python
df = data
```

```python
sns.set_theme(style = "darkgrid")
sns.displot(
    data = df.sample(frac = 0.01),
    x = "Oil_temperature",
    bins = 20
    );
```

```python
sns.set_theme(style = "whitegrid")
sns.displot(
    data = df.sample(frac = 0.01),
    x = "Oil_temperature",
    bins = 20
    );
```

```python
sns.set_theme(style = "white")
sns.displot(
    data = df.sample(frac = 0.01),
    x = "Oil_temperature",
    bins = 20
    );
```

```python
sns.set_theme(style = "dark")
sns.displot(
    data = df.sample(frac = 0.01),
    x = "Oil_temperature",
    bins = 20
    );
```

```python
sns.set_theme(style = "ticks")
sns.displot(
    data = df.sample(frac = 0.01),
    x = "Oil_temperature",
    bins = 20
    );
```

```python
sns.set_theme(style = "ticks")
sns.displot(
    data = df.sample(frac = 0.01),
    x = "Oil_temperature",
    bins = 20,
    color = "teal",
    edgecolor = "blue",
    kde = True,
    rug = True
    );
```

```python
sns.set_theme(style = "ticks")
sns.kdeplot(
    data = df.sample(frac = 0.01),
    x = "Oil_temperature"
    );
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns = ["age"])
```

```python
sns.rugplot(data = sample_ages, x = "age");
```

```python
sns.displot(
    data = sample_ages,
    x = "age", 
    rug = True,
    color = "teal",
    edgecolor = "black",
    bins = 30
    );
```

```python
sns.kdeplot(data = sample_ages, x = "age", clip = [0, 100]);
```

```python
sns.kdeplot(
    data = sample_ages, 
    x = "age", 
    clip = [0, 100], 
    bw_adjust = 0.1
    );
```

```python
sns.kdeplot(
    data = sample_ages, 
    x = "age", 
    clip = [0, 100], 
    bw_adjust = 0.6,
    fill = True
    );
```
