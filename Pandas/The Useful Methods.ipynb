```python
def dollar_signs(price):
    if price < 10:
        return "$"
    elif price <= 10 and price <= 30:
        return "$$"
    else:
        return "$$$"
```

```python
lambda num: num*2
```

```python
def tip_quality(total_bill, tip):
		if tip / total_bill > 0.25:
			return "Generous"
		else:
			return "Other" 
```

```python
tip_quality(19.00, 2)
```

```python
df["tip_generocity"] = df[["total_bill", "tip"]].apply(lambda df: tip_quality(df["total_bill"], df["tip"]), axis = 1)
```

## The Vectorize Method

```python
df["tip_generocity"] = np.vectorize(tip_quality)(df["total_bill"], df["tip"])
```
