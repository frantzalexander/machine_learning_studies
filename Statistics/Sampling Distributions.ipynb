```python
xs = np.linspace(0.55, 0.7, 100)
ys = norm.pdf(xs, p, sigma)

upper_bound = 0.65
lower_bound = 0.61

prob_less = norm.cdf(lower_bound, p, sigma)
prob_more = norm.cdf(upper_bound, p, sigma)
total_prob = prob_more - prob_less

plt.plot(
    xs,
    ys,
    label = "Prob Dist"
)

plt.axvline(
    upper_bound,
    linestyle = "--",
    color = "khaki",
    label = "Upper Bound"
    )

plt.axvline(
    lower_bound,
    linestyle = "--",
    color = "khaki",
    label = "Lower Bound"
)

plt.fill_between(
    xs,
    ys,
    where = ((xs >= lower_bound) & (xs <= upper_bound)),
    color = "red",
    alpha = 0.2,
    label = "Probability Area"
)

plt.xlabel("Probability")
plt.ylabel("Frequency")
plt.title("Probability Distribution")
plt.annotate(f"p-value: {total_prob:.4f}", (0.615, 7.5))
plt.xlim(0.56, 0.7)
plt.ylim(0, 21)
plt.legend();
```
