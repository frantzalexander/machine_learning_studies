# Legends

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 11)
```

```python
fig = plt.figure(dpi = 200)

axes = fig.add_axes([0, 0, 1,1])

axes.plot(x, x, label = "Linear")
axes.plot(x, x**2, label = "Exponential")

axes.legend();
```

```python
fig = plt.figure(dpi = 200)

axes = fig.add_axes([0, 0, 1,1])

axes.plot(x, x, label = "Linear")
axes.plot(x, x**2, label = "Exponential")

axes.legend(loc = (0.5, 0.5));
```

```python
fig = plt.figure(dpi = 200, figsize = (5, 5))

axes = fig.add_axes([0, 0, 1,1])

axes.plot(x, x, label = "Linear")
axes.plot(x, x**2, label = "Exponential")

axes.legend(loc = (1.1, 0.5));
```

# Visual Styling

```python
fig = plt.figure(figsize= (3, 3), dpi = 150)

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(
    x,
    x**2,
    label = "Linear Model",
    color = "red"
    ) 

axes.set_xlim(0, 8)
axes.set_ylim(0, 80)

axes.set_xlabel("X")
axes.set_ylabel("Y", rotation = 0)
axes.set_title("Linear Model");
```

```python
fig = plt.figure(figsize= (3, 3), dpi = 150)

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(
    x,
    x**2,
    label = "Linear Model",
    color = "#00FFFF" # RGB Hex Code
    ) 

axes.plot(
    x,
    x**2 + 10,
    label = "Linear Model +2",
    color = "#f54263"
)

axes.set_xlim(0, 8)
axes.set_ylim(0, 80)

axes.set_xlabel("X")
axes.set_ylabel("Y", rotation = 0)
axes.set_title("Linear Model")
axes.legend();
```

```python
fig = plt.figure(figsize= (3, 3), dpi = 150)

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(
    x,
    x**2,
    label = "Linear Model",
    color = "#00FFFF", # RGB Hex Code
    lw = 3,
    ls = "--"
    ) 

axes.plot(
    x,
    x**2 + 10,
    label = "Linear Model +2",
    color = "#f54263",
    linewidth = 5,
    linestyle = ":"
)

axes.set_xlim(0, 8)
axes.set_ylim(0, 80)

axes.set_xlabel("X")
axes.set_ylabel("Y", rotation = 0)
axes.set_title("Linear Model")
axes.legend();
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

lines = axes.plot(
    x, 
    x+1,
    color = "#f09f48",
    lw = 3)

lines[0].set_dashes([2,3,1,1]) ;
```

```python
fig = plt.figure(figsize= (3, 3), dpi = 150)

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(
    x,
    x,
    label = "Linear Model",
    color = "#00FFFF", # RGB Hex Code
    lw = 3,
    ls = "--"
    ) 

axes.plot(
    x,
    x + 2,
    label = "Linear Model +2",
    color = "#f54263",
    lw = 1,
    marker = "s",
    markersize = 10,
    markeredgewidth = 5,
    markeredgecolor = "green",
    markerfacecolor = "orange"
)

axes.set_xlim(0, 10)
axes.set_ylim(0, 10)

axes.set_xlabel("X")
axes.set_ylabel("Y", rotation = 0)
axes.set_title("Linear Model")
axes.legend();
```
