```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0, 8)
plt.ylim(0, 17.5)
plt.title("String Title")
plt.savefig("myfirstplot.png");
```
