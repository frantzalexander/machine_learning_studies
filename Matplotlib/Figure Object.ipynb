```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10)
y = x*2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0, 1, 1])
axes.plot(x, y);
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4 
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y);
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b);
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b);

# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x, y);
```

## Adding a Title & Labels

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B", rotation = 0)
axes1.set_title("Power of 4")

# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x, y)
axes2.set_xlim(0, 5)
axes2.set_ylim(0, 18)
axes2.set_xlabel("X")
axes2.set_ylabel("Y", rotation = 0,labelpad = 3, ha = "right")
axes2.set_title("z");
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B", rotation = 0)
axes1.set_title("Power of 4")

# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 25)
axes2.set_xlabel("A")
axes2.set_ylabel("B", rotation = 0,labelpad = 3, ha = "right")
axes2.set_title("Zoomed In");
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(a, b)
axes[1][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][1].plot(x, y)

fig.subplots_adjust(
    wspace = 1,
    hspace = 1
    );
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, dpi = 200)
axes[0][0].plot(a, b)
axes[1][0].plot(x, y)

axes[0][1].plot(a, b)
axes[0][1].set_xlim(2, 4)
axes[0][1].set_ylim(20, 500)
axes[0][1].set_xlabel("A")
axes[0][1].set_ylabel("B", rotation = 0)
axes[0][1].set_title("Power of A")


axes[1][1].plot(x, y)

fig.suptitle("This is a Figure level Title", fontsize = 16)

fig.tight_layout()
fig.savefig("My_figure2.png", bbox_inches = "tight");
```
