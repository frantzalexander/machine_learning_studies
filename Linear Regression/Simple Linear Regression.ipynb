```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/Advertising.csv")
```

```python
df.head()
```

```python
df["total_spend"] = df["TV"] + df["radio"] + df["newspaper"]
```

```python
df.head()
```

```python
sns.scatterplot(
    data = df,
    x = "total_spend",
    y = "sales",
    );
```

```python
sns.regplot(
    data = df,
    x = "total_spend",
    y = "sales",
    );
```

```python
X = df["total_spend"]
y = df["sales"]
```

```python
b1, b0= np.polyfit(X, y , deg = 1)
print(b1, b0)
```

```python
potential_spend = np.linspace(0, 500, 100)
```

```python
predicted_sales = (b1 * potential_spend) + b0 
```

```python
plt.plot(potential_spend, predicted_sales);
```

```python
sns.scatterplot(
    data = df,
    x = "total_spend",
    y = "sales"
)
plt.plot(
    potential_spend, 
    predicted_sales,
    color = 'red');
plt.xlim(0, 500)
plt.ylim(0, 30)
plt.xlabel("Total Amount Spent")
plt.ylabel("Sales")
plt.title("Total Spend vs Sales");
```

```python
spend = 200

predicted_sales = (b1 * spend) + b0
print(f"The predicted sales with a budget of {spend} would be: {predicted_sales:.2f}")
```
